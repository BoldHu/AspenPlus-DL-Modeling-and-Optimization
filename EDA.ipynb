# explore the data and find the maximum and minimum temp of 4 reactors

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# load the data
data = pd.read_csv('重整数据csv.csv')
print('load data successfully')
print(data.head())

# check the data and print the columns names
```

```python
# if the coulmn name is like '?段反应温度' and '?' means the number of the reactor
# print the column name and the minimum and maximum value of the column
# and plot the distribution of the column
temp_list = []
for i in range(1, 11):
    if '段反应温度' in data.columns[i]:
        print(data.columns[i])
        print(data[data.columns[i]].min())
        print(data[data.columns[i]].max())
        temp_list.append(data.columns[i])
# plot the four distribution of the temperature in one figure
plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)
sns.distplot(data[temp_list[0]])
plt.xlabel('Temperature of reactor 1')
plt.ylabel('density')
plt.subplot(2, 2, 2)
sns.distplot(data[temp_list[1]])
plt.xlabel('Temperature of reactor 2')
plt.subplot(2, 2, 3)
sns.distplot(data[temp_list[2]])
plt.xlabel('Temperature of reactor 3')
plt.subplot(2, 2, 4)
sns.distplot(data[temp_list[3]])
plt.xlabel('Temperature of reactor 4')
plt.show()
```

```python
# find the range of the temperature in the four reactors which can include 95% of the data
# print the range and plot the distribution of the temperature
for i in range(4):
    print(temp_list[i])
    print(data[temp_list[i]].quantile(0.025))
    print(data[temp_list[i]].quantile(0.975))
    print(data[temp_list[i]].quantile(0.975) - data[temp_list[i]].quantile(0.025))
```

```python
# find the value with most density of temperature in the four reactors
for i in range(4):
    print(temp_list[i])
    print(data[temp_list[i]].value_counts().index[0])
    print(data[temp_list[i]].value_counts().values[0])
```
